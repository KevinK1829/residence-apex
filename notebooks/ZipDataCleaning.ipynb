```python
import pandas as pd

df = pd.read_csv('https://files.zillowstatic.com/research/public_csvs/zhvi/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv?t=1780781566')
print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
print(f"Total zip codes: {df.shape[0]}")
print(f"Date columns: {df.shape[1] - 9}")
print(f"\nNull values per key column:")
print(df[['RegionName','State','City','Metro','CountyName']].isnull().sum())
print(f"\nSample of states covered:")
print(df['State'].value_counts().head(10))
```

```python
il_df = df[df['State'] == 'IL']
print(f"Illinois zip codes: {len(il_df)}")
print(f"Missing metro in IL: {il_df['Metro'].isnull().sum()}")
print(f"\nTop IL cities by zip count:")
print(il_df['City'].value_counts().head(10))
```

```python
top_metros = df[df['Metro'].notna()]['Metro'].value_counts().head(10)
print(top_metros)
```

```python
top_metro_names = top_metros.index.tolist()
metro_df = df[df['Metro'].isin(top_metro_names)].copy()

print(f"Total zip codes in top 10 metros: {len(metro_df)}")
print(f"\nNull values in date columns (sample):")
date_cols = [col for col in metro_df.columns if col.startswith('20')]
print(metro_df[date_cols].isnull().sum().describe())
```

```python
null_by_date = metro_df[date_cols].isnull().sum()
print("First 10 date columns (nulls):")
print(null_by_date.head(10))
print("\nLast 10 date columns (nulls):")
print(null_by_date.tail(10))
```

```python
null_threshold = 50
clean_dates = null_by_date[null_by_date <= null_threshold].index.tolist()
print(f"First clean date: {clean_dates[0]}")
print(f"Total clean date columns: {len(clean_dates)}")
```

```python
working_df = metro_df[['RegionID', 'RegionName', 'State', 'City', 'Metro', 'CountyName'] + clean_dates].copy()
print(f"Shape: {working_df.shape}")
print(f"Date range: {clean_dates[0]} to {clean_dates[-1]}")
print(f"Zip codes: {len(working_df)}")
working_df.head()
```

```python
working_df.to_csv('../data/Processed/metro_clean.csv', index=False)
print("Saved!")
```
